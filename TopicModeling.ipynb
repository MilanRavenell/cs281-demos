```python
id2word = gensim.corpora.Dictionary.load_from_text('simplewiki/_wordids.txt.bz2')
# load corpus iterator
mm = gensim.corpora.MmCorpus('simplewiki/_tfidf.mm')
#sliced = mm[:1000]
```

```python
print(mm)
```

```python
lda = gensim.models.ldamodel.LdaModel(corpus=mm, id2word=id2word, num_topics=100, update_every=1, chunksize=10000, passes=1)
```

```python
lda = gensim.models.ldamodel.LdaModel.load("topics.lda")
```

```python
lda.print_topics(10)
```

```python
import pyLDAvis.gensim as gensimvis
import pyLDAvis
vis_data = gensimvis.prepare(lda, mm, id2word)
```
